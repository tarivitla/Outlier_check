```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Netflix.xlsx")
df.head()
```

```python
null_values = df.isnull().sum().reset_index()
null_values
```

```python
nuls1 = null_values[null_values[0] < 1000]
nuls1
```

```python
for i in nuls1['index']:
    df.dropna(subset=[i], inplace=True)
```

```python
df['Rotten Tomatoes Score'].fillna(value=df['Rotten Tomatoes Score'].mean(), inplace=True) 
```

```python
df['Metacritic Score'].fillna(value=58.019707, inplace=True)
df['Awards Received'].fillna(value=df['Awards Received'].median(), inplace=True) 
df['Awards Nominated For'].fillna(value=df['Awards Nominated For'].median(), inplace=True) 
df['Boxoffice'].fillna(value=df['Boxoffice'].mean(), inplace=True) 
```

```python
df['Director'].fillna('unknown',inplace=True)
df['Writer'].fillna('unknown',inplace=True)
df['View Rating'].fillna('unknown',inplace=True)
df['Production House'].fillna('unknown',inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.boxplot(x=df['IMDb Votes'])
```

```python
mn_Q1 = np.percentile(df['IMDb Votes'],25)

mn_Q3 = np.percentile(df['IMDb Votes'],75)


mn_IQR = mn_Q3-mn_Q1
mn_lower_range = mn_Q1 - (1.5*mn_IQR)
mn_upper_range = mn_Q3 + (1.5*mn_IQR)

print(mn_IQR)
print(mn_lower_range)
print(mn_upper_range)
```

```python
df = df[(df['IMDb Votes'] > mn_lower_range) & (df['IMDb Votes'] < mn_upper_range)]
df.describe()
```

```python
sns.boxplot(x=df['IMDb Votes'])
```
